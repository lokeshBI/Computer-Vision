### full code

```python
# create a cascadeclassifier object
path = 'cat.2012.jpg'
path1 = "C:\\Users\\LOKESH\\Downloads\\opencv-master\\opencv-master\\data\\haarcascades\\haarcascade_frontalcatface.xml"
face_cascade = cv2.CascadeClassifier(path1)
img = cv2.imread(path) # 0 means gray scale image
grey_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(grey_image, scaleFactor = 1.05, minNeighbors = 2)
for x,y,w,h in faces:
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 3)
    plt.imshow(img)
    plt.show()
```
